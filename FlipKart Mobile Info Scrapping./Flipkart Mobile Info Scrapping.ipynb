```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
import datetime
```

```python
url = 'https://www.flipkart.com/search?q=phones%20under%2020000'
page=requests.get(url)
soup=bs(page.content,'html.parser')
#print(soup.prettify())
```

```python
names=[]
name=soup.findAll('div',class_="_4rR01T")
for i in name:
    j=i.text  
    names.append(j)
#print(names)
```

```python
ratings=[]
rating=soup.findAll('div',class_='gUuXy-')
for i in rating:
    j=i.span.div.text 
    ratings.append(j)
#print(ratings)
```

```python
l=[]
ratingcount=soup.findAll('span',class_='_2_R_DZ')
for i in ratingcount:
    j=i.span.span.text 
    l.append(j)
ratingscount = list(map(lambda x: x.replace('\xa0', ''), l))
#print(ratingscount)
```

```python
offerprices=[]
offerprice=soup.findAll('div',class_='_30jeq3 _1_WHN1')
for i in offerprice:
    j=i.text 
    offerprices.append(j)
#print(offerprices)
```

```python
originalprices=[]
originalprice=soup.findAll('div',class_='_3I9_wc _27UcVY')
for i in originalprice:
    j=i.text 
    originalprices.append(j)
#print(originalprices)
```

```python
features=[]
feature=soup.findAll('ul',class_='_1xgFaf')
for i in feature:
    j=i.text 
    features.append(j)
#print(features)
```

```python
data = {
    'Name':names,
    'Star-rating':ratings,
    'Ratings':ratingscount,
    'Offer Price':offerprices,
    'Original Price':originalprices,
    'Specifications':features
}
```

```python
df = pd.DataFrame(data)
```
